###Negative Transformation

###Logarithmic Transformation

```python
import numpy as np
```

```python
# Read the image
img = cv2.imread('/content/OIP.jpg')

# Apply logarithmic transformation
c = 255/np.log(1 + np.max(img))
log_img = c * (np.log(img + 1))

# Normalize the image to 0-255 range
log_img = np.uint8(log_img)

# Display the transformed images
plt.imshow(log_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###Gamma Transformation

```python
# Read the image
img = cv2.imread('/content/OIP.jpg')

# Apply gamma transformation
gamma = 0.5
gamma_img = np.power(img, gamma)

# Normalize the image to 0-255 range
gamma_img = np.uint8(255 * gamma_img / np.max(gamma_img))

# Display the transformed images
plt.imshow(gamma_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###Affine Transformation

```python
# Read the image
img = cv2.imread('/content/OIP.jpg')

# Define the transformation matrix
rows, cols = img.shape[:2]
src_points = np.float32([[0, 0], [cols-1, 0], [0, rows-1]])
dst_points = np.float32([[cols-1, 0], [0, 0], [cols-1, rows-1]])
affine_matrix = cv2.getAffineTransform(src_points, dst_points)

# Apply affine transformation
affine_img = cv2.warpAffine(img, affine_matrix, (cols, rows))

# Display the transformed images
plt.imshow(affine_img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

###Cropping Transformation

```python
# Read the image
img = cv2.imread('/content/OIP.jpg')

# Define the region of interest (ROI)
x, y, w, h = 100, 100, 200, 200
roi = img[y:y+h, x:x+w]

# Display the cropped images
plt.imshow(roi)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
