##TRANSLATION

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img=cv2.imread('/content/drive/MyDrive/OIP.jpg')
```

```python
rows,cols=img.shape[:2]
```

```python
translation_matrix=np.float32([[1,0,70],[0,1,110]])
```

```python
print(translation_matrix)
```

```python
img_translation=cv2.warpAffine(img,translation_matrix,(cols,rows))
```

```python
img_translation=cv2.warpAffine(img,translation_matrix,(cols+240,rows+120))
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img_translation)
```

##ROTATION

```python
height,width=img.shape[:2]
```

```python
center=(width/2,height/2)
```

```python
rotate_matrix=cv2.getRotationMatrix2D(center=center,angle=45,scale=1)
```

```python
rotate_image=cv2.warpAffine(src=img,M=rotate_matrix,dsize=(width,height))
```

```python
cv2_imshow(img)
```

```python
cv2_imshow(rotate_image)
```

```python
cv2.waitKey(0)
```

##SCALING

```python
img=cv2.imread('/content/drive/MyDrive/OIP.jpg')
```

```python
img_scaled = cv2.resize(img,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_LINEAR)
cv2_imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img,None,fx=1.2, fy=1.2, interpolation = cv2.INTER_CUBIC)
cv2_imshow(img_scaled)
```

```python
img_scaled = cv2.resize(img,(450, 400), interpolation = cv2.INTER_AREA)
cv2_imshow(img_scaled)
```

```python
cv2.waitKey()
```

##SHEARING


```python
M = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img = cv2.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
cv2_imshow(sheared_img)
cv2.waitKey(0)
```

```python
M = np.float32([[1,   0, 0], [0.5, 1, 0], [0,   0, 1]])
sheared_img = cv2.warpPerspective(img, M, (int(cols*1.5), int(rows*1.5)))
cv2_imshow(sheared_img)
cv2.waitKey(0)
```

##REFLECTION

```python
M = np.float32([[1,  0, 0],
                [0, -1, rows],
                [0,  0, 1]])
reflected_img = cv2.warpPerspective(img, M,
                                   (int(cols),
                                    int(rows)))
cv2_imshow(reflected_img)
cv2.imwrite('reflection_out.jpg', reflected_img)
cv2.waitKey(0)
```
